#### 2. pc

```python
dataset['pc'].value_counts()
```

```python
dataset['pc'] = dataset['pc'].replace(to_replace = {'normal' : 0, 'abnormal' : 1})
```

#### 3. pcc

```python
dataset['pcc'].value_counts()
```

```python
dataset['pcc'] = dataset['pcc'].replace(to_replace = {'notpresent':0,'present':1})
```

#### 4. ba

```python
dataset['ba'].value_counts()
```

```python
dataset['ba'] = dataset['ba'].replace(to_replace = {'notpresent':0,'present':1})
```

#### 5. htn

```python
dataset['htn'].value_counts()
```

```python
dataset['htn'] = dataset['htn'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

#### 6. dm

```python
dataset['dm'].value_counts()
```

```python
dataset['dm'] = dataset['dm'].replace(to_replace = {'\tyes':'yes', ' yes':'yes', '\tno':'no'})
```

```python
dataset['dm'] = dataset['dm'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

#### 7. cad

```python
dataset['cad'].value_counts()
```

```python
dataset['cad'] = dataset['cad'].replace(to_replace = {'\tno':'no'})
```

```python
dataset['cad'] = dataset['cad'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

#### 8. appet

```python
dataset['appet'].unique()
```

```python
dataset['appet'] = dataset['appet'].replace(to_replace={'good':1,'poor':0,'no':np.nan})
```

#### 9. pe

```python
dataset['pe'].value_counts()
```

```python
dataset['pe'] = dataset['pe'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

#### 10. ane

```python
dataset['ane'].value_counts()
```

```python
dataset['ane'] = dataset['ane'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

#### 11. classification

### Converting Object values into Numeric values:

```python
dataset['pcv'] = pd.to_numeric(dataset['pcv'], errors='coerce')
dataset['wc'] = pd.to_numeric(dataset['wc'], errors='coerce')
dataset['rc'] = pd.to_numeric(dataset['rc'], errors='coerce')
```

### Handling Null Values:

- There is Outliers present in our dataset so We fill NaN values with Median.

```python
dataset.columns
```

```python
features = ['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr', 'bu',
           'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad',
           'appet', 'pe', 'ane']
```

```python
for feature in features:
    dataset[feature] = dataset[feature].fillna(dataset[feature].median())
```

```python
# Independent and Dependent Feature:
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X.head()
```

```python
# Feature Importance: determine and visualize the feature importance using the ExtraTreesClassifier algorithm from the scikit-learn library.
from sklearn.ensemble import ExtraTreesClassifier #ExtraTreesClassifier is an ensemble learning algorithm based on decision trees that can be used for classification tasks.
import matplotlib.pyplot as plt 
model=ExtraTreesClassifier() #This line creates an instance of the ExtraTreesClassifier model.
model.fit(X,y) #This line fits the ExtraTreesClassifier model to the input features X and the corresponding target variable y. It trains the model on the provided data.

plt.figure(figsize=(8,6))
ranked_features=pd.Series(model.feature_importances_,index=X.columns) #this line creates a pandas Series object called ranked_features that contains the feature importances calculated by the ExtraTreesClassifier model. 
ranked_features.nlargest(24).plot(kind='barh') #selecting the best 24 features
plt.show()
```

```python
X = dataset[['sg', 'htn', 'hemo', 'dm', 'al', 'appet', 'rc', 'pc']]
X.head()
```

```python
X.tail() #rows
```

```python
y.head() 
```

```python
# Train Test Split:# splitting for learning task 3% for test 70% train
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=33)
```

```python
print(X_train.shape)
print(X_test.shape)
```

#### There is no need of Standardization and Normalization of our dataset, as we using Ensemble Technique.
Ensemble techniques are used in machine learning to improve predictive performance by combining the predictions of multiple individual models

```python
# Importing Performance Metrics: A confusion matrix provides a tabular representation of the predicted labels versus the true labels, allowing you to evaluate the model's performance in terms of true positives, true negatives, false positives, and false negatives.
#This line imports the classification_report function, which generates a text-based report that includes several evaluation metrics for each class in a classification task

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# RandomForestClassifier:
from sklearn.ensemble import RandomForestClassifier #This line imports the RandomForestClassifier class from the ensemble module in scikit-learn. The Random Forest Classifier is an ensemble learning method based on decision trees.
RandomForest = RandomForestClassifier() #
RandomForest = RandomForest.fit(X_train,y_train)

# Predictions:
y_pred = RandomForest.predict(X_test) #The predict() function predicts the target variable labels based on the input features.

# Performance:
print('Accuracy:', accuracy_score(y_test,y_pred)) 
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# AdaBoostClassifier:AdaBoost (Adaptive Boosting) is an ensemble learning method that combines weak learners (typically decision trees) to create a strong classifier.
from sklearn.ensemble import AdaBoostClassifier 
AdaBoost = AdaBoostClassifier()
AdaBoost = AdaBoost.fit(X_train,y_train)

# Predictions:
y_pred = AdaBoost.predict(X_test)

# Performance:
print('Accuracy:', accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# GradientBoostingClassifier: Gradient Boosting is an ensemble learning method that combines multiple weak learners 
from sklearn.ensemble import GradientBoostingClassifier
GradientBoost = GradientBoostingClassifier()
GradientBoost = GradientBoost.fit(X_train,y_train)

# Predictions:
y_pred = GradientBoost.predict(X_test)

# Performance:
print('Accuracy:', accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
